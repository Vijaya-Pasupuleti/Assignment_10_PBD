```python
# Question 3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split

url_titanic = "C:/Users/vijay/OneDrive/Desktop/Assignment_10_PBD/titanic.csv"
df = pd.read_csv(url_titanic)
print(df.head())

# Check for missing values
print(df.isnull().sum())

# Age, Cabin and Embarked have missing values
# Handling missing values in Age column by filling the empty rows with mean of age 
df['Age'].fillna(df['Age'].mean(), inplace=True)

# Handling missing values by dropping the categorical variables
df.dropna(inplace=True)
print(df.head(10),"\n")

# Identify outliers using box plots
numer_cols = df.select_dtypes(include=np.number).columns

plt.figure(figsize=(12, 6))
sns.boxplot(data=df[numer_cols], orient='h')
plt.xticks(rotation=45)
plt.show()

#Age, Fare have outliers
#Removing outliers using 1.5 IQR method

Q1 = df[numer_cols].quantile(0.25)
Q3 = df[numer_cols].quantile(0.75)
IQR = Q3 - Q1

remove_out = df[~((df[numer_cols] < (Q1 - 1.5 * IQR)) | (df[numer_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]

print(remove_out)

plt.figure(figsize=(12, 6))
sns.boxplot(data=remove_out[numer_cols], orient='h')
plt.xticks(rotation=45)
plt.show()

# Feature Engineering
# Scaling the numberical columns from int to float using MinMaxScaler

scaler = MinMaxScaler()
numeric_cols = ['Age', 'Survived', 'Pclass', 'SibSp','Parch', 'Fare']
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
print(df.head(10))

# Question 4
# Separate the features (X) (choosing numerical columns) and target (y)( choosing Categorical variable)
X = df[numeric_cols]
y = df['Sex']

# Spliting the data into 70 percent training and 30 percent testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Choosing RandomForest and Support Vector Machine algorithms for analysis
# Initializing the models
initialize_rf = RandomForestClassifier(random_state=42)
initialize_svm = SVC(random_state=42)

# Training the models
initialize_rf.fit(X_train, y_train)
initialize_svm.fit(X_train, y_train)

# Testing the models
prediction_randfor = initialize_rf.predict(X_test)
prediction_supportvec = initialize_svm.predict(X_test)

# Evaluating the models based on various metrics like accuracy, precision, recall and f1-score to compare the performance
def evaluation_rf_svm(predictions, model_name):
    accuracy = accuracy_score(y_test, predictions)
    report = classification_report(y_test, predictions)
    print(f"Accuracy of {model_name} model: {accuracy:.2f}")
    print(f"Classification Report for {model_name} Model:\n{report}")

evaluation_rf_svm(prediction_randfor, "Random Forest")
evaluation_rf_svm(prediction_supportvec, "Support Vector Machine")

#Question 5
# Visualization using pairplot by importing seaborn and matplotlib
sns.pairplot(df,vars=['Sex'], palette='deep') 
plt.show()
```
